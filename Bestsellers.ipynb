```python
import pandas as pd
```

```python
df = pd.read_csv('bestsellers.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.rename(columns={"Name": "Title", "Year": "Publication Year", "User Rating": "Rating"}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Price'] = df['Price'].astype("float")
```

```python
df.info()
```

#Analysing 'Author' popularity

```python
author_counts = df['Author'].value_counts()
author_counts
```

# Average rating by Genre


```python
avg_rate_by_genre = df.groupby('Genre')['Rating'].mean()
avg_rate_by_genre
```

Exporting the insights

```python
# Export top selling authors to a CSV file
author_counts.head(10).to_csv("top_authors.csv")

# Export average rating by genre to a CSV file
avg_rate_by_genre.to_csv("avg_rating_by_genre.csv")
```
